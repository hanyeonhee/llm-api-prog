```python
from openai import OpenAI
client = OpenAI()

response = client.embeddings.create(
    input="왕",
    model="text-embedding-3-small"
)

print(response.data[0].embedding)
```

```python
def embedding(input):
    response = client.embeddings.create(
        input=input,
        model="text-embedding-3-small"
    )

    return response.data[0].embedding
```

## 코사인 유사도

```python
import numpy as np

def cosine_similarity(a, b):
    return np.dot(a, b) / (np.linalg.norm(a) * np.linalg.norm(b))
```

### 단어의 코사인 유사도

```python
print(f"'여왕'과 '여자'의 코사인 유사도: {cosine_similarity(embedding('여왕'), embedding('여자'))}")
```

```python
import pandas as pd

df = pd.DataFrame({'단어': ['공부', '글', '도서', '문서', '영화', '책상', '화장품']})
df['embedding'] = df['단어'].apply(embedding)
df
```

```python
# df의 각 단어에 대해 '책'과의 코사인 유사도를 계산하고 내림차순 정렬

df['cosine_similarity'] = df['embedding'].apply(lambda x: cosine_similarity(x, embedding('책')))
df.sort_values(by='cosine_similarity', ascending=False)
```

```python
# df의 각 단어에 대해 질의문과 코사인 유사도를 계산하고 내림차순 정렬

df['cosine_similarity'] = df['embedding'].apply(
    lambda x: cosine_similarity(x, embedding('아니 땐 굴뚝에 연기 나랴?'))
)
df.sort_values(by='cosine_similarity', ascending=False)
```
